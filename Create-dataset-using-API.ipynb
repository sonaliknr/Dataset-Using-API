```python
import numpy as np
import pandas as pd
import requests
import json
```

```python
url = "https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1"
JSONContent = requests.get(url).json()
content = json.dumps(JSONContent, indent = 4, sort_keys=True)
print(content)
```

```python
df=[]
for i in range(10):
    url = "https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page="
    JSONContent = requests.get(url+str(i+1)).json()
    result=JSONContent['results']
    for item in result:
        df.append([item['id'],item['title'],item['overview'],item['release_date'],item['popularity'],item['vote_average'],item['vote_count']])
```

```python
data=pd.DataFrame(df)
```

```python
data.columns=['Id','Title','Overview','Release_Date','Popularity','Vote_Average','Vote_Count']
```
